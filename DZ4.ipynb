Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

3.1. Исследуем зависимость цены от количества жилых метров в доме. 

```python
g = sns.JointGrid(data=df, x="sqft_living", y="price",height=10)
g.plot_joint(sns.scatterplot, s=100, alpha=.5)
g.plot_marginals(sns.histplot, kde=True);
```

Вывод: зависисомть цены от размера жилой площади прямо проворциональная, чем выше жилая площадь дома, тем выше цена за дом. 

3.2. Исследуем зависимость цены от года постройки дома. 

```python
df5= df.groupby('yr_built')['price'].mean().reset_index()
df5
```

```python
plt.figure(figsize=(17, 8))
plt.scatter(df5['yr_built'], df5['price'])
plt.title('Зависимость цены от года постройки дома');
```

Вывод: дома постренные в 1900 -е  годы и в нулевые годы имею высокую цену, это означает что у покупателей большой интерес вызывают не только новые дома, но и построенные столетие назад, т.е. раритетные дома. Низкий порог цен наблюдается на дома построенные в период с 1960 по 1970  годы и в период Второй мировой войны. 

3.3. Исследуем зависимость цены от количества этажей в доме. 

```python
plt.figure(figsize=(10,6))
sns.boxplot(x=df['price'], y=df['floors'].astype('str'), whis=1.5)
plt.title('Взаимосвязь стоимости домов и количества этажей')
plt.xlabel('Цена')
plt.ylabel('Этажность дома')
```

Самый маленький разброс по ценам имеют дома с 3,5 этажами. Самый большой разброс по цене имеют дома с 2,5 этажами. Дома с 2-мя этажами имеют самое большое количество выбросов по цене, дома с 3,5 этажами имеют наименьшее количество выбросов. Самые дешевые дома имеют 1 этаж. 


3.4 Рассмотрим зависмость цены на дом от состояния дома. 

```python
plt.figure(figsize=(10,6))
sns.lineplot(x=df['condition'], y=df['price'])
plt.title('Взаимосвязь стоимости домов и состояния')
plt.xlabel('Состояние')
plt.ylabel('Цена')
```

```python
sns.jointplot(x = df['price'], y = df['condition'], kind = 'reg', height=12, ratio=4);
```

На графике видно что цена практически напрямую зависит от состояния дома, чем лучше состояние дома тем выше цена. Наибольшее количесво домой имеет удовлетворительное состояние. 

3.5. Рассмотрим зависимость цены от класса дома.  

```python
plt.figure(figsize=(10,6))
sns.boxplot(x=df['grade'], y=df['price'], whis=1.5)
plt.title('Зависимость стоимости дома от класса')
plt.xlabel('Класс')
plt.ylabel('Этажность дома')
```

Из графика видно, что цена прямопропорционально зависит от стоимости дома, чем выше класс дома, тем выше его соимость. 

3.6. Исследуем зависимость цены дома от наличия набережной.

```python
plt.figure(figsize=(10,6))
sns.boxplot(x=df['price'], y=df['waterfront'].astype('str'), whis=1.5, hue=df['grade'])
plt.title('Pависимость цены дома от наличия набережной')
plt.xlabel('Цена')
plt.ylabel('Линия набережной')
```

Из графика видно что разброс цен на дома в зависимости от наличия набережной достаточно большой, стоимость 50 % предлагаемых домов зависит от наличия набережной. Наблюдается большое поличество выбросов у домой удаленных от набережной.
